```python
# SIMULADOR DE IA: ASISTENTE DE RELOJERO Y LUPA
# ---
# Créditos iniciales (se ejecutan una sola vez fuera del ciclo)

print("Bienvenido al Cerebro Lógico del Relojero Digital")

# Variable de control para el menú
opcion_menu = "0"

# Estructura repetitiva para mantener el sistema activo
while opcion_menu != "3":
    print("\n--- MENÚ DE IA PARA RELOJERÍA ---")
    print("1. Evaluar Reloj (Asistente de Diagnóstico)")
    print("2. Gestionar Lupa y Mesa (Organizador de Taller)")
    print("3. Apagar Sistema")

    opcion_menu = input("\nSeleccione una opción (1-3): ")

    # MODO 1: Evaluación inteligente de relojes
    if opcion_menu == "1":
        print("\n[MODO IA: EVALUADOR DE RELOJES]")
        marca = input("Marca o modelo del reloj: ")
        complejidad = int(input("Nivel de complejidad del mecanismo (1 al 10): "))
        antiguedad_anos = int(input("Años transcurridos desde su fabricación: "))
        es_original = input("¿El mecanismo cuenta con piezas originales? (si/no): ").lower()

        # Lógica de IA para decidir el destino y cuidado del reloj
        if complejidad >= 8 and antiguedad_anos >= 50 and es_original == "si":
            # Operador lógico 'and' para piezas de alta relojería o museo
            print(f"IA: ¡ALERTA! '{marca}' es una PIEZA HISTÓRICA / ALTA RELOJERÍA.")
            print("IA: Recomendación: Manipular exclusivamente con guantes de seda y herramientas de titanio.")
        elif complejidad > 5 or es_original == "si":
            # Operador 'or' para reparaciones estándar pero valiosas
            print(f"IA: '{marca}' es un Reloj Mecánico Estándar.")
            print("IA: Recomendación: Realizar mantenimiento general y lubricación de rubíes.")
        else:
            print(f"IA: '{marca}' es un mecanismo simple o réplica.")
            print("IA: Recomendación: Reparación rápida o cambio directo de batería/cuarzo.")

    # MODO 2: Organización inteligente del taller y nivel de aumento (Lupa)
    elif opcion_menu == "2":
        print("\n[MODO IA: ORGANIZADOR DE TALLER Y LUPA]")
        cantidad_engranajes = int(input("¿Cuántos engranajes/piezas sueltas vas a revisar?: "))
        luz_ambiente = float(input("Nivel de iluminación en tu banco de trabajo (0.0 a 100.0 lúmenes): "))
        tipo_trabajo = input("¿El trabajo actual es de 'limpieza' o 'calibracion'?: ").lower()

        # Lógica para decidir el tipo de lupa y soporte
        if tipo_trabajo == "calibracion" and luz_ambiente > 20.0:
            print("IA: Perfil detectado: MICRO-CALIBRACIÓN DE ESCAPE.")
            # Operación aritmética simple para estimar el aumento de lupa requerido (X)
            aumento_lupa = cantidad_engranajes * 2
            print(f"IA: Usar 'Lupa Monocular de Ojo'. Aumento recomendado: {aumento_lupa}X.")
        elif cantidad_engranajes > 15 or luz_ambiente < 10.0:
            # Si hay poca luz o demasiadas micro-piezas, se necesita el microscopio electrónico
            print("IA: Perfil detectado: RECONSTRUCCIÓN COMPLEJA.")
            print("IA: Recomendación: Usar MICROSCOPIO DIGITAL con luz LED propia para evitar fatiga visual.")
        else:
            print("IA: Perfil detectado: INSPECCIÓN RÁPIDA.")
            print("IA: Recomendación: Usar Lupa de Visera (Soporte de diadema) básica.")

    # MODO 3: Salida
    elif opcion_menu == "3":
        print("\nCerrando sesión de IA... ¡Buen trabajo, maestro relojero!")

    # Manejo de opciones no válidas
    else:
        print("\nIA: Error. Opción no reconocida por el sistema del taller.")
```
